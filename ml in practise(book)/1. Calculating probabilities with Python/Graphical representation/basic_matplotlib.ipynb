```python
import matplotlib.pyplot as plt
```

```python
# Графическая отрисовка линейного соотношения
x = range(0,10)
y = [2 * value for value in x]
plt.plot(x,y)
plt.show()
```

```python
# Отрисовка отдельных точек данных
plt.scatter(x, y)
plt.show()
```

```python
# Затенение интервала под отрисованным графиком
plt.plot(x, y)

def is_in_interval(number, minimum, maximum):
    return minimum <= number <= maximum
    
where = [is_in_interval(value, 2, 6) for value in x]
plt.fill_between(x,y, where=where)
plt.show()
```

```python
# Выражение отдельных координат на непрерывном графике

plt.scatter(x, y)
plt.plot(x, y)
plt.fill_between(x, y, where=where)
plt.show()
```

```python
# Добавление подписей осей
plt.plot(x, y)
plt.xlabel('Values between zero and ten')
plt.ylabel('Twice the values of x')
plt.show()
```
